```python
allNewLines = []

with open('australian_users_items.json', 'r', encoding='utf-8') as file1:
    contenido = file1.read()
    
    for line in contenido.split('\n'):
        if line.strip():
            NewDict1 = ast.literal_eval(line)
            allNewLines.append(NewDict1)

UserItemsDF = pd.DataFrame(allNewLines)
```

```python
allNewLines = []

with open('australian_user_reviews.json', 'r', encoding='utf-8') as file2:
    contenido = file2.read()
    
    for line in contenido.split('\n'):
        if line.strip():
            NewDict2 = ast.literal_eval(line)
            allNewLines.append(NewDict2)

UserReviewsDF = pd.DataFrame(allNewLines)
```

```python
all_data = []

with open('output_steam_games.json', 'r', encoding='utf-8') as file3:
    for line in file3:
        try:
            data = json.loads(line)
            all_data.append(data)
        except json.JSONDecodeError as e:
            print(f"Error al decodificar JSON en l√≠nea: {line}")
            print(f"Mensaje de error: {str(e)}")

SteamGamesDF = pd.DataFrame(all_data)
```
